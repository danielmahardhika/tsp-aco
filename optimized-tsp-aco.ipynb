```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk mengimpor data dari file teks dan mengabaikan header atau metadata
def impor_data(nama_file):
    kota = []
    with open(nama_file, 'r') as file:
        for baris in file.readlines():
            # Mengabaikan baris yang berisi metadata
            if baris.startswith("NAME") or baris.startswith("TYPE") or baris.startswith("COMMENT") or baris.startswith("EDGE_WEIGHT_TYPE"):
                continue
            # Mengabaikan baris kosong
            if not baris.strip():
                continue
            # Mengambil data koordinat kota
            data = baris.split()
            try:
                id_kota = int(data[0])  # ID kota
                x = float(data[1])       # Koordinat X
                y = float(data[2])       # Koordinat Y
                kota.append((id_kota, x, y))
            except ValueError:
                continue
    return kota

class AdvancedACO:
    def __init__(self, kota, max_evaluasi=5000, jumlah_semut=50, evaluasi_per_run=50, 
                 alpha=1, beta=5, rho=0.1, q0=0.9, elitism_factor=2.0):
        self.kota = kota
        self.max_evaluasi = max_evaluasi
        self.jumlah_semut = jumlah_semut
        self.evaluasi_per_run = evaluasi_per_run
        self.alpha = alpha
        self.beta = beta
        self.rho = rho
        self.q0 = q0
        self.elitism_factor = elitism_factor
        
        # Matriks jarak dan pheromone
        self.jarak = self.hitung_matriks_jarak()
        self.pheromone = np.ones((len(kota), len(kota))) * 0.1
        
        # Variabel untuk melacak solusi terbaik
        self.solusi_terbaik = []
        self.jarak_terbaik = float('inf')
        
        # Parameter penyesuaian dinamis
        self.pheromone_min = 0.0001
        self.pheromone_max = 1.0

    def hitung_matriks_jarak(self):
        n = len(self.kota)
        matriks_jarak = np.zeros((n, n))
        for i in range(n):
            for j in range(n):
                matriks_jarak[i][j] = self.hitung_jarak(self.kota[i], self.kota[j])
        return matriks_jarak

    def hitung_jarak(self, kota1, kota2):
        return np.sqrt((kota1[1] - kota2[1])**2 + (kota1[2] - kota2[2])**2)

    def local_search_swap(self, jalur):
        """Implementasi local search sederhana dengan swap"""
        improved = True
        while improved:
            improved = False
            for i in range(len(jalur)):
                for j in range(i+1, len(jalur)):
                    # Lakukan swap
                    jalur_baru = jalur.copy()
                    jalur_baru[i], jalur_baru[j] = jalur_baru[j], jalur_baru[i]
                    
                    panjang_lama = self.hitung_total_jarak(jalur)
                    panjang_baru = self.hitung_total_jarak(jalur_baru)
                    
                    if panjang_baru < panjang_lama:
                        jalur = jalur_baru
                        improved = True
                        break
                if improved:
                    break
        return jalur

    def hitung_total_jarak(self, jalur):
        total_jarak = 0
        for i in range(len(jalur)):
            total_jarak += self.jarak[jalur[i]][jalur[(i+1) % len(jalur)]]
        return total_jarak

    def konstruksi_jalur_semut(self):
        jalur_semut = []
        for _ in range(self.jumlah_semut):
            jalur = self.konstruksi_jalur_tunggal()
            jalur_semut.append(jalur)
        return jalur_semut

    def konstruksi_jalur_tunggal(self):
        jalur = []
        dikunjungi = [False] * len(self.kota)
        
        # Mulai dari kota acak
        kota_sekarang = np.random.randint(len(self.kota))
        jalur.append(kota_sekarang)
        dikunjungi[kota_sekarang] = True
        
        while len(jalur) < len(self.kota):
            kota_selanjutnya = self.pilih_kota_selanjutnya(kota_sekarang, dikunjungi)
            jalur.append(kota_selanjutnya)
            dikunjungi[kota_selanjutnya] = True
            kota_sekarang = kota_selanjutnya
        
        return jalur

    def pilih_kota_selanjutnya(self, kota_sekarang, dikunjungi):
        kota_kandidat = [i for i in range(len(self.kota)) if not dikunjungi[i]]
        
        # Probabilitas untuk setiap kota kandidat
        probabilitas = []
        for kota in kota_kandidat:
            tau = self.pheromone[kota_sekarang][kota] ** self.alpha
            eta = (1.0 / self.jarak[kota_sekarang][kota]) ** self.beta
            probabilitas.append(tau * eta)
        
        # Normalisasi probabilitas
        total = sum(probabilitas)
        probabilitas = [p / total for p in probabilitas]
        
        # Pilih kota berdasarkan probabilitas
        return kota_kandidat[np.argmax(probabilitas)]

    def update_pheromone(self, jalur_semut):
        # Penguapan pheromone
        self.pheromone *= (1 - self.rho)
        
        # Deposit pheromone untuk setiap jalur
        for jalur in jalur_semut:
            panjang_jalur = self.hitung_total_jarak(jalur)
            deposit = self.elitism_factor / panjang_jalur
            
            for i in range(len(jalur)):
                kota_asal = jalur[i]
                kota_tujuan = jalur[(i+1) % len(jalur)]
                self.pheromone[kota_asal][kota_tujuan] += deposit
                self.pheromone[kota_tujuan][kota_asal] += deposit
        
        # Batasi pheromone
        self.pheromone = np.clip(self.pheromone, self.pheromone_min, self.pheromone_max)

    def jalankan_aco(self):
        evaluasi = 0
        hasil_jarak = []
        
        while evaluasi < self.max_evaluasi:
            # Konstruksi jalur semut
            jalur_semut = self.konstruksi_jalur_semut()
            
            # Local search untuk setiap jalur
            jalur_semut_optimized = [self.local_search_swap(jalur) for jalur in jalur_semut]
            
            # Update pheromone
            self.update_pheromone(jalur_semut_optimized)
            
            # Temukan solusi terbaik
            for jalur in jalur_semut_optimized:
                jarak_jalur = self.hitung_total_jarak(jalur)
                if jarak_jalur < self.jarak_terbaik:
                    self.jarak_terbaik = jarak_jalur
                    self.solusi_terbaik = jalur
            
            # Catat hasil
            hasil_jarak.append(self.jarak_terbaik)
            evaluasi += self.evaluasi_per_run
            
            print(f"Evaluasi: {evaluasi}, Jarak Terbaik: {self.jarak_terbaik}")
        
        return self.jarak_terbaik, np.mean(hasil_jarak), np.std(hasil_jarak), hasil_jarak

def visualisasi_titik(kota, judul):
    x = [t[1] for t in kota]
    y = [t[2] for t in kota]

    plt.figure(figsize=(12, 10))
    plt.scatter(x, y, c='red', marker='o', s=100, label="Kota")

    for i, t in enumerate(kota):
        plt.annotate(str(t[0]), (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9)

    plt.title(f"Visualisasi Titik Kota - {judul}", fontsize=16)
    plt.xlabel("Koordinat X", fontsize=12)
    plt.ylabel("Koordinat Y", fontsize=12)
    plt.grid(True)
    plt.legend()
    plt.show()

def visualisasi_titik_dan_jalur(kota, solusi_terbaik, judul):
    x = [t[1] for t in kota]
    y = [t[2] for t in kota]

    plt.figure(figsize=(14, 12))
    plt.scatter(x, y, c='red', marker='o', s=100, label="Kota")

    for i, t in enumerate(kota):
        plt.annotate(str(t[0]), (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9)

    # Konversi solusi terbaik ke indeks yang sesuai
    solusi_terbaik_adjusted = [idx - 1 for idx in solusi_terbaik]

    # Buat jalur
    jalur_x = [kota[idx][1] for idx in solusi_terbaik_adjusted]
    jalur_y = [kota[idx][2] for idx in solusi_terbaik_adjusted]
    
    # Sambungkan kembali ke titik awal
    jalur_x.append(kota[solusi_terbaik_adjusted[0]][1])
    jalur_y.append(kota[solusi_terbaik_adjusted[0]][2])

    plt.plot(jalur_x, jalur_y, color='blue', linewidth=2, label="Jalur Terbaik")

    plt.title(f"Visualisasi Titik dan Jalur Terbaik - {judul}", fontsize=16)
    plt.xlabel("Koordinat X", fontsize=12)
    plt.ylabel("Koordinat Y", fontsize=12)
    plt.grid(True)
    plt.legend()
    plt.show()

def visualisasi_konvergensi(hasil_jarak_terbaik, judul):
    plt.figure(figsize=(12, 6))
    plt.plot(range(0, 5000, 50), hasil_jarak_terbaik, marker='o')
    plt.title(f"Kurva Konvergensi - {judul}", fontsize=16)
    plt.xlabel("Evaluasi", fontsize=12)
    plt.ylabel("Jarak Terbaik", fontsize=12)
    plt.grid(True)
    plt.show()

def eksperimen_multiple(kota, judul, run_count=10):
    semua_hasil = []
    solusi_terbaik_seluruh_run = []

    for run in range(run_count):
        print(f"\n--- Percobaan {run + 1} ---")
        aco = AdvancedACO(kota)
        terbaik, rata_rata, std_deviasi, hasil_jarak_terbaik = aco.jalankan_aco()
        
        semua_hasil.append((terbaik, rata_rata, std_deviasi, hasil_jarak_terbaik))
        solusi_terbaik_seluruh_run.append(aco.solusi_terbaik)

        # Visualisasi konvergensi untuk setiap run
        visualisasi_konvergensi(hasil_jarak_terbaik, f"{judul} - Run {run+1}")

    # Statistik akhir
    terbaik_total = min([hasil[0] for hasil in semua_hasil])
    rata_rata_total = np.mean([hasil[1] for hasil in semua_hasil])
    std_deviasi_total = np.std([hasil[2] for hasil in semua_hasil])

    # Temukan solusi terbaik
    indeks_solusi_terbaik = np.argmin([hasil[0] for hasil in semua_hasil])
    solusi_terbaik_akhir = solusi_terbaik_seluruh_run[indeks_solusi_terbaik]
    solusi_terbaik_akhir_adjusted = [idx + 1 for idx in solusi_terbaik_akhir]

    print(f"\nHasil Akhir {judul}:")
    print(f"Jarak Terbaik: {terbaik_total}")
    print(f"Rata-rata Jarak: {rata_rata_total}")
    print(f"Standar Deviasi: {std_deviasi_total}")
    print(f"Jalur Terbaik: {solusi_terbaik_akhir_adjusted}")

    return solusi_terbaik_akhir_adjusted, terbaik_total, rata_rata_total, std_deviasi_total, semua_hasil

# Impor dan proses dataset
kota_berlin52 = impor_data('berlin52.txt')
#kota_kroa100 = impor_data('kroa100.txt') uncomment kalo mau dijalankan
#kota_ch150 = impor_data('ch150.txt') uncomment kalo mau dijalankan

# Visualisasi titik kota
visualisasi_titik(kota_berlin52, "Berlin52")
#visualisasi_titik(kota_kroa100, "KroA100") uncomment kalo mau dijalankan
#visualisasi_titik(kota_ch150, "Ch150") uncomment kalo mau dijalankan

# Eksperimen untuk setiap dataset
solusi_berlin52, terbaik_berlin52, rata_rata_berlin52, std_berlin52, _ = eksperimen_multiple(kota_berlin52, "Berlin52")
#solusi_kroa100, terbaik_kroa100, rata_rata_kroa100, std_kroa100, _ = eksperimen_multiple(kota_kroa100, "KroA100") uncomment kalo mau dijalankan
#solusi_ch150, terbaik_ch150, rata_rata_ch150, std_ch150, _ = eksperimen_multiple(kota_ch150, "Ch150") uncomment kalo mau dijalankan

# Visualisasi jalur terbaik
visualisasi_titik_dan_jalur(kota_berlin52, solusi_berlin52, "Berlin52")
#visualisasi_titik_dan_jalur(kota_kroa100, solusi_kroa100, "KroA100") uncomment kalo mau dijalankan
#visualisasi_titik_dan_jalur(kota_ch150, solusi_ch150, "Ch150") uncomment kalo mau dijalankan

```
